Class Code

```python
class Student:
    def __init__(self, name, id, gender=None, dob=None):
        if name.count(" ") > 0:
            print("Name added")
            self.name = name
        else:
            print("Invalid Name Given")
            pass
        if len(str(id)) == 8:
            print("Student ID added")
            self.id = id
        else:
            print("Student ID must have 8 digits")
            pass
        self.gender = gender
        self.dob = dob
        if gender is None:
            self.gender = "Unknown"
        if dob is None:
            self.dob = "Unknown"
        self.uniqueKey = name+"-"+id
        
        self.semesters_completed = []                                           # takes info from Course class
        self.semester_details = {}
        if name.count(" ") > 0:
            self.semester_details[self.name] = {}

    def studentDetails(self):
        print(f"Name: {self.name}\nID: {self.id}\nGender: {self.gender}\nDate of Birth: {self.dob}")


class Course:
    def __init__(self, c_name, c_code, credit_count):
        self.c_name = c_name
        self.c_code = c_code
        self.credit_count = credit_count
        self.course_details = {}
        self.student_semester_details = {}
        self.c_student_li = []
        self.s_semester_li = []

    def AddStudent(self, student, semester):        
        self.c_student_li.append(student.name)
        self.student_name = student.name
        self.course_details[self.c_code] = self.c_student_li
        
        if semester not in student.semesters_completed:
            student.semesters_completed.append(semester)
        self.s_semester_li = student.semesters_completed

        if semester not in student.semester_details[student.name]:
            student.semester_details[student.name][semester] = [self.c_code]
        else:
            student.semester_details[student.name][semester].append(self.c_code)

        self.student_semester_details[self.c_code] = {}
        self.student_semester_details[self.c_code][student.name] = semester
    
    def courseDetails(self):
        print(f"Course Name: {self.c_name}")
        print(f"Course Code: {self.c_code}")
        print(f"Credit Count: {self.credit_count}")
        print(f"{self.c_code} course details for {self.student_name}: {self.student_semester_details[self.c_code]}")
        print(f"{self.student_name}'s completed semester(s): {', '.join(self.s_semester_li)}")
```
