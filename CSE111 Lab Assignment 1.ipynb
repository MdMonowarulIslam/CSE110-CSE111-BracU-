````python
# Task 1

s = input("Enter your string: ")
uc, lc = 0, 0
for i in range(len(s)):
    if s[i].isupper():
        uc += 1
    else:
        lc += 1

if lc > uc:
    print(s.lower())
elif uc > lc:
    print(s.upper())
else:
    print(s.lower())
````

````python
# Task 2

s = input("Enter your input: ")
ch, dg = 0, 0
ans = ""
for x in s:
    if x.isdigit():
        dg += 1
    elif x.isalpha():
        ch += 1
if dg == 0:
    ans += "WORD"
elif ch == 0:
    ans += "NUMBER"
else:
    ans += "MIXED"

print(ans)
````

````python
# Task 3

s = input("Enter your string: ")
ucl = ""
p1, p2 = 0, 0

for x in s:
    if x.isupper():
        ucl += x

p1, p2 = s.find(ucl[0]), s.find(ucl[1])
if p1-p2 > 1:
    ans = s[p1 + 1:p2]
else:
    ans = "BLANK"

print(ans)
````

````python
# Task 4

s = input("Enter string: ")
ns = ""

li = s.split(",")
a, b = li[0], li[1]

for x in a:
    if x in b:
        ns += x
for x in b:
    if x in a:
        ns += x

if ns == "":
    ns = "Nothing in common"

print(ns)
````

````python
# Task 5

s = input("Enter your Password: ")
lc, uc, dg, spc = 0, 0, 0, 0

for x in s:
    if x.isupper():
        uc += 1
    elif x.islower():
        lc += 1
    elif x.isdigit():
        dg += 1
    else:
        spc += 1

ans = []
if dg == 0:
    ans.append("Digit missing")
if lc == 0:
    ans.append("Lowercase character missing")
if uc == 0:
    ans.append("Uppercase character missing")
if spc == 0:
    ans.append("Special character missing")

if len(ans) == 0:
    print("OK")
else:
    print(*[ans[i] for i in range(len(ans))], sep=", ")
````

````python
# Task 1

li = []
while True:
    x = input("Enter your element: ")
    if x.upper() == "STOP":
        break
    else:
        li.append(int(x))

n_li = []    
for x in li:
    if x not in n_li:
        n_li.append(x)

for x in n_li:
    counter = 0
    for y in li:
        if x == y:
            counter += 1
    print(x, "-", counter, "times")
````

```Dictionary and Tuples```

````python
# Task 1

di1 = {"a": 100, "b": 100, "c": 200, "d": 300}
di2 = {"a": 300, "b": 200, "d": 400, "e": 200}

di3 = di1.copy()
di3.update(di2)

for k1, v1 in di1.items():
    for k2, v2 in di2.items():
        if k1 == k2:
            di3[k1] = (v1 + v2)
print(di3)

v_li = []
for v in di3:
    v_li.append(di3[v])
for x in v_li:
    if v_li.count(x) > 1:
        v_li.remove(x)
v_li.sort()
values = tuple(v_li)
print("Values:", values)
````

````python
# Task 1

def BMI_calc(h, w):
    bmi = w/h**2
    x = ""
    if bmi < 18.5:
        x = "Underweight"
    elif 18.5 <= bmi <= 24.9:
        x = "Normal"
    elif 25 <= bmi <= 30:
        x = "Overweight"
    elif bmi > 30:
        x = "Obese"

    print("Score is", str(bmi)+". You are", x)


height = int(input("Enter your height in CMs: ")) / 100
weight = int(input("Enter your weight in KGs: "))
BMI_calc(height, weight)
````

````python
# Task 2

def func_t2(mn, mx, di):
    ans = 0
    for i in range(mn, mx):
        if i % di == 0:
            ans += i
    print(ans)


x = int(input("Enter the minimum value: "))
y = int(input("Enter the maximum value: "))
z = int(input("Enter the divisor: "))

func_t2(x, y, z)
````

````python
# Task 4

def replace_domain(m, n_d, o_d=None):
    mail_name = m[0: m.find("@") + 1]
    mail_domain = m[m.find("@") + 1: len(m)+1]

    if mail_domain != n_d:
        print("Changed:", mail_name+n_d)
    else:
        print("Unchanged:", m)


mail = input("enter the mail address: ")
new_domain = input("enter the new domain: ")
old_domain = input("enter the old domain: ")

replace_domain(mail, new_domain, old_domain)
replace_domain("alice@kaaj.com", "sheba.xyz", "kaaj.com")
replace_domain("bob@sheba.xyz", "sheba.xyz")
````

````python
# Task 5

def palindrome_finder(x):
    nx = ""
    for letters in x:
        if letters != " ":
            nx += letters

    if nx == nx[::-1]:
        print("Palindrome")
    else:
        print("Not a Palindrome")


s = input("Enter your string: ").upper()
palindrome_finder(s)
````
