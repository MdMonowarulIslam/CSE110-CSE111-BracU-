```python
# Task 1

class Calculator:
    def __init__(self):
        print("Let's calculate!")

    def add(self, x, y):
        return x + y

    def subtract(self, x, y):
        return x - y

    def multiply(self, x, y):
        return x*y

    def divide(self, x, y):
        return x/y


calc = Calculator()

v1 = int(input("Value 1: "))
op = input("Operator: ")
v2 = int(input("Value 2: "))

ans = 0
if op == "+":
    ans = calc.add(v1, v2)
elif op == "-":
    ans = calc.subtract(v1, v2)
elif op == "*":
    ans = calc.multiply(v1, v2)
elif op == "/":
    ans = calc.divide(v1, v2)

print("Result:", ans)
```

```python
# Task 5

class Shape:
    def __init__(self, shape, p1, p2):
        self.shape = shape
        self.parameter1 = p1
        self.parameter2 = p2


    def area(self):
        if self.shape.upper() == "TRIANGLE" or self.shape.upper() == "RHOMBUS":
            area = 1/2*self.parameter1*self.parameter2
        elif self.shape.upper() == "RECTANGLE" or self.shape.upper() == "SQUARE":
            area = self.parameter1*self.parameter2
        else:
            area = "Shape unknown"
        print("Area:", area)

    
triangle = Shape("Triangle",10,25)
triangle.area()
print("==========================")
square = Shape("Square",10,10)
square.area()
print("==========================")
rhombus = Shape("Rhombus",18,25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle",15,30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium",15,30)
trapezium.area()
```

```python
# Task 7

class Student:
    def __init__(self, name, id, dept, grades):
        self.name = name
        self.id = id
        self.department = dept
        self.grades = grades
    
    def calculate_CGPA(self):
        self.CGPA = sum(self.grades)/len(self.grades)
    
    def print_details(self):
        print("Name:", self.name, "ID:", self.id)
        print("Department:", self.department)
        print("CGPA:", format(self.CGPA, ".2f"))
        if self.CGPA > 3.80:
            x = "'Highest Distinction'"
        elif self.CGPA > 3.65:
            x = "'High Distinction'"
        elif self.CGPA > 3.50:
            x = "'Distinction'"
        elif self.CGPA >= 2.00:
            x = "'Satisfactory'"
        elif self.CGPA < 2.00:
            print("Sorry, you cannot graduate")
            return
        
        print("Your academic standing is", x)


s1 = Student('Dora', '15995599','CSE', [4,3.7,3.7,4])
s1.calculate_CGPA()
print("==========================")
s1.print_details()
print("==========================")
s2 = Student('Pingu', '12312322', 'EEE', [1.7,1.3,1.3,1.3,1])
s2.calculate_CGPA()
print("==========================")
s2.print_details()
print("==========================")
s3 = Student('Bob', '13311331', 'CSE', [2,3,3,3.7,2.7,2.7])
s3.calculate_CGPA()
print("==========================")
s3.print_details()
```

```python
# Task 8

class Shinobi:
    def __init__(self, name, rank):
        self.name = name
        self.rank = rank
        self.missions = 0
        self.salary = 0
        self.salary_multiplier = 0

    def calSalary(self, missions):
        if self.rank == "Genin":
            self.salary_multiplier = 50
        elif self.rank == "Chunin":
            self.salary_multiplier = 100
        else:
            self.salary_multiplier = 500
        
        self.missions = missions
        self.salary = missions*self.salary_multiplier

    def changeRank(self, new_rank):
        self.rank = new_rank
        
    def printInfo(self):
        print("Name:", self.name)
        print("Rank:", self.rank)
        print("Number of missions:", self.missions)
        print("Salary:", self.salary)


naruto = Shinobi("Naruto", "Genin")
naruto.calSalary(5)
naruto.printInfo()
print('====================')
shikamaru = Shinobi('Shikamaru', "Genin")
shikamaru.printInfo()
shikamaru.changeRank("Chunin")
shikamaru.calSalary(10)
shikamaru.printInfo()
print('====================')
neiji = Shinobi("Neiji", "Jonin")
neiji.calSalary(5)
neiji.printInfo()
```

```python
# Task 10

class UberEats:
    def __init__(self, name, phone, address):
        self.name = name
        self.phone = phone
        self.address = address
        self.orders = {}
        self.price = 0
        print(name+", welcome to UberEats!")

    def add_items(self, item1, item2, price1, price2):
        self.orders = {item1: price1, item2: price2}
        self.price = price1 + price2

    def print_order_detail(self):
        details = "User details: Name:"+ self.name +", Phone: "+str(self.phone)+", Address: "+self.address+"\nOrders: "+str(self.orders)+"\nTotal Paid Amount: "+str(self.price)
        return details


order1 = UberEats("Shakib", "01719658xxx", "Mohakhali")
print("=========================")
order1.add_items("Burger", "Coca Cola", 220, 50)
print("=========================")
print(order1.print_order_detail())
print("=========================")
order2 = UberEats ("Siam", "01719659xxx", "Uttara")
print("=========================")
order2.add_items("Pineapple", "Dairy Milk", 80, 70)
print("=========================")
print(order2.print_order_detail())
```

```python
# Task 11

class Spotify:
    def __init__(self, li):
        self.spotify_li = li
        print("Welcome to Spotify!")
    
    def playing_number(self, index):
        try:
            played_song = "Playing "+ str(index)+" number song for you"+"\nSong name:"+self.spotify_li[index-1]
        except IndexError:
            played_song = str(index)+" number song not found. Your playlist has "+str(len(self.spotify_li))+" songs only"
        return played_song + "\n########################"

    def add_to_playlist(self, song):
        self.spotify_li.append(song)


user1 = Spotify(["See You Again", "Uptown Funk", "Hello"])
print("=========================")
print(user1.playing_number(4))
user1.add_to_playlist("Dusk Till Dawn")
print(user1.playing_number(3))
print(user1.playing_number(4))
```
