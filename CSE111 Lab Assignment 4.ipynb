```python
# Task 7

class Match:
    def __init__(self, teams):
        self.t1 = teams[:teams.find("-")]
        self.t2 = teams[teams.find("-")+1:]
        print("5..4..3..2..1.. Play !!!")
        self.runs = 0
        self.overs = 0
        self.wickets = 0

    def add_run(self, runs=0):
        self.runs += runs
    
    def add_over(self, overs=0):
        if self.overs + overs > 5:
            print(f"Warning! Cannot add {overs} over/s (5 over match)")
        else:
            self.overs += overs
    
    def add_wicket(self, wickets=0):
        self.wickets += wickets

    def print_scoreboard(self):
        scoreBoard = f"Batting Team: {self.t1}\nBowling Team {self.t2}\nTotal Runs: {self.runs} Wickets: {self.wickets} Overs: {self.overs}"
        return scoreBoard

match1 = Match("Rangpur Riders-Cumilla Victorians")
print("=========================")
match1.add_run(4)
match1.add_run(6)
match1.add_over(1)
print(match1.print_scoreboard())
print("=========================")
match1.add_over(5)
print("=========================")
match1.add_wicket(1)
print(match1.print_scoreboard())
```

```python
# Task 13

class Account:
    def __init__(self, name=None, balance=None):
        self.name = name
        if name is None:
            self.name = "Default Account"        
        if balance is None:
            self.balance = 0.0
        else:
            self.balance = float(balance)

    def details(self):
        detail = f"{self.name}\n{self.balance}"
        return detail

    def withdraw(self, wdraw_amount):
        x = (self.balance - wdraw_amount)
        if x > 0 and wdraw_amount < self.balance*0.6930:
            print("Withdraw successful! New balance is:", x)
            self.balance = x
        else:
            print("Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum.")
        

a1 = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930)
print("------------------------")
a2.withdraw(600)
print("------------------------")
a1.withdraw(6929)
```

```python
# Task 14

class StudentDatabase:
    def __init__(self, name, id):
        self.name = name
        self.id = id
        self.grades = {}
    
    def calculateGPA(self, grades, semester):
        course_li, grade_li = [], []
        for x in grades:
            c_n = x[:x.find(":")]
            c_g = float(x[x.find(":")+2:])
            course_li.append(c_n)
            grade_li.append(c_g)
        courses = tuple(course_li)
        gpa = sum(grade_li) / len(grade_li)
        self.grades[semester] = {}
        self.grades[semester][courses] = ("%.2f" % gpa)
    
    def printDetails(self):
        print(f"Name: {self.name}")
        print(f"ID: {self.id}")
        for sem in self.grades:
            print(f"Course taken in {sem}:")
            for courses in self.grades[sem]:
                for x in courses:
                    print(x)
                print(self.grades[sem][courses])

s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```

```python
# Task 21

class Student:
    def __init__(self, name, id, dept):
        self.name = name
        self.id = id
        self.dept = dept

    def details(self):
        detail = f"Name: {self.name}\nID: {self.id}\nDepartment: {self.dept}"
        return detail
    
    def advise(self, *courses):
        credits = float(len(courses)*3)
        if credits < 9.0:
            status = "You have to take at least 1 more course."
        elif credits > 12.0:
            status = "You have to drop at least 1 course."
        else:
            status = "Ok"
        print(f"{self.name}, you habe taken {credits} credits.")
        print(f"List of courses: {', '.join(courses)}")
        print(f"Status: {status}")

    
s1 = Student("Alice","20103012","CSE")
s2 = Student("Bob", "18301254","EEE")
s3 = Student("Carol", "17101238","CSE")
print("##########################")
print(s1.details())
print("##########################")
print(s2.details())
print("##########################")
s1.advise("CSE110", "MAT110", "PHY111")
print("##########################")
s2.advise("BUS101", "MAT120")
print("##########################")
s3.advise("MAT110", "PHY111", "ENG102", "CSE111", "CSE230")
```

```python
# Task 23

class Author:
    def __init__(self, name=None):
        self.name = name
        self.books = {}
        self.book_number = 0
    
    def setName(self, name):
        self.name = name

    def addBook(self, bname, bgenre):
        if self.name == None:
            print("A book can not be added without author name")
        else:
            self.book_number += 1
            if bgenre not in self.books:
                self.books[bgenre] = [bname]
            else:
                self.books[bgenre].append(bname)
    
    def printDetail(self):
        print(f"Number of Book(s): {self.book_number}")
        print(f"Author Name: {self.name}")
        for key in self.books:
            print(f"{key}: {', '.join(self.books[key])}")


a1 = Author()
print("=================================")
a1.addBook("Ice", "Science Fiction")
print("=================================")
a1.setName("Anna Kavan")
a1.addBook("Ice", "Science Fiction")
a1.printDetail()
print("=================================")
a2 = Author("Humayun Ahmed")
a2.addBook("Onnobhubon", "Science Fiction")
a2.addBook("Megher Upor Bari", "Horror")
print("=================================")
a2.printDetail()
a2.addBook("Ireena", "Science Fiction")
print("=================================")
a2.printDetail()
print("=================================")
```

```python
# Task 25

class Vaccine:
    def __init__(self, v_name, v_origin, delay):
        self.v_name = v_name
        self.v_origin = v_origin
        self.delay = delay
    
class Person:
    def __init__(self, name, age, prof=None):
        self.name = name
        self.age = age
        self.prof = prof
        if prof == None:
            self.prof = "General Citizen"
        self.vacc = ""
        self.doses = {}
        self.vacc_li = []
    
    def pushVaccine(self, vacc, dose="1st Dose"):
        self.vacc = vacc.v_name
        self.vacc_li.append(vacc.v_name)

        if self.vacc == self.vacc_li[0]:
            if self.age < 25 and self.prof != "Student":
                print(f"Sorry {self.name}, Minimum age for taking vaccines is 25 years now.")
            else:
                print(f"{dose} done for {self.name}")
                if dose == "1st Dose":
                    self.doses[dose] = "Given"
                    self.doses["2nd Dose"] = f"Please come after {vacc.delay} days."
                else:
                    self.doses["1st Dose"] = "Given"
                    self.doses[dose] = "Given"
        else:
            print(f"Sorry {self.name}, you can't take 2 different vaccines")
    
    def showDetail(self):
        print(f"Name: {self.name} Age: {self.age} Type: {self.prof}")
        print(f"Vaccine Name: {self.vacc}")
        for key in self.doses:
            print(f"{key}: {self.doses[key]}")


astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person("Bob", 21, "Student")
print("=================================")
p1.pushVaccine(astra)
print("=================================")
p1.showDetail()
print("=================================")
p1.pushVaccine(sin, "2nd Dose")
print("=================================")
p1.pushVaccine(astra, "2nd Dose")
print("=================================")
p1.showDetail()
print("=================================")
p2 = Person("Carol", 23, "Actor")
print("=================================")
p2.pushVaccine(sin)
print("=================================")
p3 = Person("David", 34)
print("=================================")
p3.pushVaccine(modr)
print("=================================")
p3.showDetail()
print("=================================")
p3.pushVaccine(modr, "2nd Dose")
```
