```python
# Task 1

class PlayerEarning:
    def __init__(self, name):
        self.name = name
        self.earning = 0
        self.bonus = 0
        self.total_earning = 0
    
    def calculateTotal(self, earning, goals=0):
        self.earning = earning
        if goals > 30:
            self.bonus = (5/100)*earning + 10000
        elif 30 > goals > 0:
            self.bonus = (5/100)*earning
        self.total_earning = earning + self.bonus
    
    def printDetails(self):
        print(f"Player Name: {self.name}")
        print(f"Player Season Earning without bonus: {self.earning}")
        print(f"Bonus: {int(self.bonus)}")
        print(f"Player Season Earning After Bonus: {self.total_earning}")

print("**********************")
player1 = PlayerEarning('Buffon')
player1.calculateTotal(250000)
player1.printDetails()
print("\n**********************")
player2 = PlayerEarning('Dybala')
player2.calculateTotal(250000, 31)
player2.printDetails()
print("\n**********************")
player3 = PlayerEarning('Cuadrado')
player3.calculateTotal(250000, 20)
player3.printDetails()
```

```python
class myList:
    def init(self, *args):
        self.lst = list(args)
        self.total = 0
        self.average = 0

    def merge(self,*m):
        self.lst2 = list(m)
        for j in self.lst2:
            self.lst.append(j)
        if len(self.lst2) > 0:
            self.total = 0

    def sum(self):
        for i in self.lst:
            self.total += i
        print(f'Sum: {self.total}')

    def average(self):
        self.average = self.total / len(self.lst)
        print(f'Average: {self.average}')

l1 =  myList(2,3,4,5,6)
l1.sum()
l1.merge(4,5,9)
l1.sum()
l1.average()
print("-----------------------------")
l2 =  myList()
l2.average()
l2.merge(1,2,4,8)
l2.sum()
```

```python
# Task 10

class Student:
    def __init__(self,name,ID):
        self.name = name
        self.ID = ID
    def Details(self):
        return "Name: "+self.name+"\n"+"ID: "+self.ID+"\n"

class CSEStudent(Student):
    def __init__(self, name, ID, semester):
        super().__init__(name, ID)
        self.semester = semester
        self.course_cg = {}
    
    def Details(self):
        return f"Name: {self.name}\nID: {self.ID}\nCurrent Semester: {self.semester}"

    def addCourseWithMarks(self, *courses_marks):
        for i in range(0, len(courses_marks), 2):
            self.course_cg[courses_marks[i]] = courses_marks[i+1]
        
        for key in self.course_cg:
            if self.course_cg[key] >= 85:
                self.course_cg[key] = 4.0
            elif 80 <= self.course_cg[key] <= 84:
                self.course_cg[key] = 3.3
            elif 70 <= self.course_cg[key] <= 79:
                self.course_cg[key] = 3.0
            elif 65 <= self.course_cg[key] <= 69:
                self.course_cg[key] = 2.3
            elif 57 <= self.course_cg[key] <= 64:
                self.course_cg[key] = 2.0
            elif 55 <= self.course_cg[key] <= 56:
                self.course_cg[key] = 1.3
            elif 50 <= self.course_cg[key] <= 54:
                self.course_cg[key] = 1.0
            elif self.course_cg[key] < 50:
                self.course_cg[key] = 0.0
            
    def showGPA(self):
        print(f"{self.name} has taken {len(self.course_cg)} courses.")
        for key in self.course_cg:
            print(f"{key}: {self.course_cg[key]}")
        print(f"GPA of {self.name} is {sum(self.course_cg.values())/len(self.course_cg):.2f}")


Bob = CSEStudent("Bob","20301018","Fall 2020")
Carol = CSEStudent("Carol","16301814","Fall 2020")
Anny = CSEStudent("Anny","18201234","Fall 2020")
print("#########################")
print(Bob.Details())
print("#########################")
print(Carol.Details())
print("#########################")
print(Anny.Details())
print("#########################")
Bob.addCourseWithMarks("CSE111",83.5,"CSE230",73.0,"CSE260",92.5)
Carol.addCourseWithMarks("CSE470",62.5,"CSE422",69.0,"CSE460",76.5,"CSE461",87.0)
Anny.addCourseWithMarks("CSE340",45.5,"CSE321",95.0,"CSE370",91.0)
print("----------------------------")
Bob.showGPA()
print("----------------------------")
Carol.showGPA()
print("----------------------------")
Anny.showGPA()
```

```python
# Quiz 4

class Product:
    def __init__(self, name, quantity, price):
        self.name = name
        self.quantity = quantity
        self.__actual_price = price*quantity
    
    def get_price(self):
        return self.__actual_price
    
    def showInfo(self):
        print(f"Name of the product: {self.name}")
        print(f"Quantity: {self.quantity}")
        print(f"Total Price: {self.get_price()}")

class Sale_Product(Product):
    def __init__(self, discount, name, quantity, price):
        super().__init__(name, quantity, price)
        self.discount = discount
        self.price_sale = self.get_price() - (discount/100)*self.get_price()
    
    def showInfo(self):
        super().showInfo()
        print(f"Discount Rate: {self.discount}")
        print(f"Price after sale: {self.price_sale}")

class Regular_Product(Product):
    pass

class Specialist_Sale_Product(Sale_Product):
    __FeatureList = []
    def __init__(self, discount, name, quantity, price, req_f, price_dict):
        super().__init__(discount, name, quantity, price)
        self.req = req_f
        self.price_dict = price_dict
    
    @classmethod
    def setFeatureList(cls, li):
        cls.__FeatureList = li
    
    def showComparison(self):
        print(f"Config of {self.name}:")
        stock_features = []
        stock_models = {}
        total_price = 0

        for key, value in self.price_dict.items():
            stock_features.append(key[(key.index("_")+1):])

            for elem in value:
                stock_models[elem[0]] = elem[1]
        
        for req in self.req:
            if req[0] not in Specialist_Sale_Product.__FeatureList:
                print(f"Could not add {req[0]} as not in feature list")
            else:
                if req[0].lower() in stock_features:
                    if req[1] in stock_models.keys():
                        print(f"Added {req[0]}. Price= {stock_models[req[1]]}")
                        total_price += stock_models[req[1]]
                    else:
                        print(f"Could not add {req[0]} as model not in stock")
                else:
                    print(f"Could not add {req[0]} as not currently in stock")
        
        total_price *= self.quantity
        print(f"Total Price = {total_price}")
        print(f"You paid {self.price_sale}")

        if total_price > self.price_sale:
            print("It's a steal!")
        else:
            print("Fooled you :(")        

        
General_product = Product("Soap", 5, 10)
General_product.showInfo()
print("------------------")
Discounted_soap = Sale_Product(15, "Dove", 6, 20)
Discounted_soap.showInfo()
print("------------------")
Normal_soap = Regular_Product("Lux", 2, 5)
print(Normal_soap.get_price())
Normal_soap.showInfo()
print("------------------")
feature_list= ['GPU', 'CPU', 'RAM', 'MOBO']

Specialist_Sale_Product.setFeatureList(feature_list)

My_required_features = (('MONITOR', 'IPS'),('CPU', 'inteli9'), ('RAM', 8), ('GPU', 'RTX2080'))

price_dictionary = {'latest_cpu': [['inteli9', 100]],  'available_gpu':[['RTX3090', 500], ['RTX3080', 350]]}

PC = Specialist_Sale_Product(20, "New Rig", 1, 100, My_required_features, price_dictionary)
PC.showComparison()
print("------------------")
My_required_features = (('CPU', 'inteli9'), ('GPU', 'RTX3090'))

OtherPC = Specialist_Sale_Product(15, "Another Rig", 2, 850, My_required_features, price_dictionary)
OtherPC.showComparison()
```
