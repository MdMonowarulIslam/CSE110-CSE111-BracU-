```python
class Student:
    def __init__(self, name, id, age, gender, *args):
        self.name = name
        self.id = id
        self.age = age
        self.gender = gender
        self.courses = []
        self.vehicles = []
        
        if len(args) == 4:
            self.courses = [list(args[:5])]
        elif len(args) == 7:
            self.courses = [list(args[:5])]
            self.vehicles = [list(args[4:])]

    def AddCourse(self, *course_details):
        self.courses.append(list(course_details))

    def AddVehicle(self, *vehicle_details):
        self.vehicles.append(list(vehicle_details))
    
    def getName(self):
        self.name = input("Enter name: ")
    
    def setName(self, name):
        self.name = name

    def printDetails(self):
        print(f"Name: {self.name}")
        print(f"ID: {self.id}")
        print(f"Age: {self.age}")
        print(f"Gender: {self.gender}")
        if len(self.courses) == 0:
            print("No Courses completed! New Student")
        else:
            for x in self.courses:
                print("-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_")
                print(f"Course name: {x[0]}")
                print(f"Course code: {x[1]}")
                print(f"Grades achieved: {x[2]}")
                print(f"Semester completed: {x[3]}")
                print("-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_")
        if len(self.vehicles) == 0:
            print("Does not use vehicles")
        else:
            for x in self.vehicles:
                print("-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_")
                print(f"Vehicle plate number: {x[0]}")
                print(f"Vehicle type: {x[1]}")
                print(f"Vehicle color: {x[2]}")
                print("-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_")

        

s1 = Student("Mike", 213, 20, "Male")
s1.printDetails()
print("================================")
s1.getName()
s1.printDetails()
print("================================")
s1.setName("Mike")
s1.printDetails()
print("================================")
s1.AddCourse("Programming Language 2", "CSE111", 3.7, 3)
s1.printDetails()
print("================================")
s1.AddVehicle("DHA-CHA-221", "SUV", "Black")
s1.printDetails()
print("================================")

s2 = Student("Riley", 213, 20, "Female", "Design 3", "ARC201", 4.0, 3)
s2.printDetails()
print("================================")

s3 = Student("Riley", 213, 20, "Female", "Design 3", "ARC201", 4.0, 3, "DHA-KA-202", "Scooty", "Blue")
s3.printDetails()
print("================================")
```
